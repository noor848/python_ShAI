#### 5. Read the total profit of each month and show it using the histogram to see the most common profit ranges.  
**Note:**It must be implemented in three methods.

```python
####################################### histogram Plot using Matplotlib visualization #########################################
```

```python
# Creating histogram
profit_range=[150000, 175000, 200000, 225000, 250000, 300000, 350000]
plt.hist(df['total_profit'],profit_range)
plt.xticks(np.arange(150000,350000,25000))
plt.yticks(np.arange(1,6,1))
plt.xticks([150000, 175000, 200000, 225000, 250000, 300000, 350000]
)
plt.legend(loc='upper left',labels=['profit data'])
plt.xlabel("profit range in dollar ")
plt.ylabel("Actual profit in dollar ")
plt.title("Profit data")
plt.show()
```

#### 6. Calculate total sale data for last year for each product and show it using a Pie chart.
- **Note:**
    - In Pie chart display Number of units sold per year for each product in percentage.
    - It must be implemented in two methods.

The Pie chart should look like this.


```python
#################################### pie Plot using Matplotlib visualization #####################################
```

```python
#total_units

labels=['facecream','facewash','toothpaste','Bathingsoap','shampoo','moisturizer']
sizes = [df['facecream'].sum(), df['facewash'].sum(), df['toothpaste'].sum(), df['bathingsoap'].sum(),df['shampoo'].sum(),df['moisturizer'].sum()]

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        )
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(labels=labels,loc='lower right')
plt.show()
```

```python
#################################### pie Plot using Pandas visualization #####################################
```

```python


plot = pd.DataFrame(sizes,index=labels).plot(kind='pie', figsize=(5, 5),subplots=True,autopct='%1.1f%%')
plt.legend(labels=labels,loc='lower right')
plt.show()
```
